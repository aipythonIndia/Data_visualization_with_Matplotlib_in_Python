### Slicing the required data for plotting

```python
high_val = data['High']
high_val
```

```python
low_val = data['Low']
low_val
```

```python
high_val.shape, low_val.shape
```

```python
%matplotlib notebook
plt.plot(high_val,'v--r' ,label = 'Highest Val')
plt.plot(low_val, label = 'Lowest Val')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('EURO vs INR price', fontsize =14)
plt.grid()
plt.legend()
```

```python
open_val = data['Open']
```

```python
%matplotlib notebook
plt.plot(high_val,'v--r' ,label = 'Highest Val')
plt.plot(low_val, label = 'Lowest Val')
plt.plot(open_val, label = 'Open Val')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('EURO vs INR price', fontsize =14)
plt.grid()
plt.legend()
```

```python
%matplotlib notebook
adj_close = data['Adj Close']

plt.plot(high_val,'v--r' ,label = 'Highest Val')
plt.plot(low_val, label = 'Lowest Val')
plt.plot(open_val, label = 'Open Val')
plt.plot(adj_close, label = 'Ajusted close')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('EURO vs INR price', fontsize =14)
plt.grid()
plt.legend()
```
